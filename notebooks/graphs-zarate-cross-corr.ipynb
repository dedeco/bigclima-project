<h3>Correlação cruzada</h3>

<p>Considerando a variável alvo não há nenhuma correlação significativa linearmente com outras variáveis. Estamos considerando todos os registros da região sudeste.</p>

```python
dfaux = df
#dfaux = dfaux[(dfaux['wsnm']=='BELO HOR. (PAMPULHA)')]
len(dfaux)
```

```python
TARGET = ['prcp',]
OTHERS = ['wsid','stp','smax','smin','temp','tmax','tmin','dewp','dmax','dmin'\
         ,'hmdy','hmax','hmin','wdsp','wdct','gust']
```

```python
dfaux.index = pd.to_datetime(dfaux.mdct)
for v in dfaux.columns:
    if v not in TARGET + OTHERS:
        dfaux = dfaux.drop(v,1)
```

```python
dfaux.head(4)
```

```python
def df_shifted(df, target=None, lag=0):
    if not lag and not target:
        return df       
    new = {}
    for c in df.columns:
        if c == target:
            new[c] = df[target]
        else:
            new[c] = df[c].shift(periods=lag)
    return  pd.DataFrame(data=new)
```

```python
dfshif = df_shifted(dfaux,'prcp',lag=-1)
```

```python
order_list = ['wsid','prcp','stp','smax','smin','temp','tmax','tmin',\
              'dewp','dmax','dmin','hmdy','hmax','hmin','wdsp','wdct','gust']
dfshif = dfshif.reindex(order_list, axis=1)
```

```python
dfshif.head(15)
```

```python
dfaux.head(15)
```

```python
lag ={}
for i in range(1,25):
    dfshif = df_shifted(dfaux,'prcp',lag=-i)
    have_corr = []
    d = {}
    coef = dfshif.astype(float).corr()
    for k,v in coef.prcp.items():
        if v > 0.7 or v< -0.7:
            if k != 'prcp':
                d[k] = v
                if d:
                    have_corr.append(d)    
    if have_corr:
        lag[i] = have_corr
```

```python
print (lag)
```

```python
dfshif = df_shifted(dfaux,'prcp',lag=-3)
order_list = ['prcp','stp','temp','dewp','hmdy','wdsp','wdct','gust']
dfshif = dfshif.reindex(order_list, axis=1)
coef = dfshif.astype(float).corr()
```

```python
print len(dfshif)
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(15,10))
plt.title(u'3 horas', y=1.05, size=16)

mask = np.zeros_like(coef)
mask[np.triu_indices_from(mask)] = True

svm = sns.heatmap(coef, mask=mask, linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)

fig = svm.get_figure()
name = '../graphs-br/pearson-base-all_3_hora.png'
fig.savefig(name,dpi=200)
```
